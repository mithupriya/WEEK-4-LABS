```r
names(Boston)
```

```r
# fit a simple linear regression
lm.fit=lm(medv~lstat ,data=Boston)
attach(Boston)
lm.fit=lm(medv~lstat)
lm.fit
```

```r
# find more information
summary(lm.fit)
```

```r
names(lm.fit)
```

```r
coef(lm.fit)
```

```r
# to obatin confidence interval for coefficient estimates
confint(lm.fit)
```

```r
#to produce confidence  interval for prediction of medv for lstat
predict(lm.fit,data.frame(lstat=c(5,10,15)) , interval="confidence")
```

```r
#to produce cprediction  interval for prediction of medv for lstat

predict(lm.fit,data.frame(lstat=c(5,10,15)) , interval="prediction")
```

```r
plot(lstat,medv)
abline(lm.fit)
```

```r
plot(lstat,medv)
abline(lm.fit,lwd=3)
abline(lm.fit,lwd=3,col="red")
plot(lstat,medv,col="red")
plot(lstat,medv,pch=20)
plot(lstat,medv,pch="+")
plot(1:20,1:20,pch=1:20)
```

```r
par(mfrow=c(2,2))
plot(lm.fit)
```

```r
 plot(predict (lm.fit), residuals (lm.fit))
 plot(predict (lm.fit), rstudent (lm.fit))
```

```r
#compute levearge statistics and index of the largest element of vector
plot(hatvalues (lm.fit ))
 which.max (hatvalues (lm.fit))
```

```r
# multiple regression
lm.fit=lm(medv~lstat+age,data=Boston )
summary(lm.fit)
```

```r
 lm.fit =lm(medv~.,data=Boston) 
 summary (lm.fit)
```

```r
# data from another dataset
library(car)
vif(lm.fit)
```

```r
# regression using all predictors except age
lm.fit1=lm(medv~.-age ,data=Boston) 
 summary (lm.fit1)
```

```r
 lm.fit1=update (lm.fit , ~.-age)
```

```r
# include interaction terms in linear model
summary (lm(medv~lstat *age ,data=Boston ))
```

```r
# regression of medv on lstat and lstat^2
lm.fit2=lm(medv~lstat +I(lstat ^2))
summary(lm.fit2)
```

```r
 lm.fit =lm(medv~lstat)
 anova(lm.fit ,lm.fit2)
```

```r
 par(mfrow=c(2,2))
plot(lm.fit2)
```

```r
# produce 5th order polynomial fit
lm.fit5=lm(medv~poly(lstat ,5))
summary(lm.fit5)
```

```r
# try log transformation
summary (lm(medv~log(rm),data=Boston ))
```
